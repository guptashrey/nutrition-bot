```python
from scidownl import scihub_download
import pandas as pd
```

```python
doi_df = pd.read_pickle("./doi.pkl")  
doi_df
```

```python
for doi in doi_df['doi']:
    paper = f"https://doi.org/{doi}"
    paper_type = "doi"
    out = f'./pdf/{doi.replace("/", "_")}.pdf'
    try:
        scihub_download(paper, paper_type=paper_type, out=out)
    except:
        pass
```
